# Arima Model

```python
#Import the librairy needed
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from pandas.plotting import register_matplotlib_converters 
register_matplotlib_converters()
from sklearn.metrics import mean_squared_error
```

Download the Data and creat the log og the inflation index and the logarithmic growth

```python
#Download the data which are on my drive
from google.colab import drive
drive.mount('/content/drive')
serie_inflation = pd.read_csv('/content/drive/My Drive/Copie de CPIAUCSL.csv')
#we create a data frame with only the datas we will use for the model 
serie_inflation["DATE"]=pd.to_datetime(serie_inflation["DATE"])
serie_inflation.set_index('DATE',inplace=True)

#Creating the Log and le Log differenciating log
serie_inflation["Log Inflation"] = np.log(serie_inflation["CPIAUCSL"])
serie_inflation["Log shifting"] = serie_inflation["Log Inflation"] - serie_inflation["Log Inflation"].shift()
serie_inflation["Log shifting"][0] = 0
plt.figsize=(15,7)
plt.plot(serie_inflation)
```

Splitting the data into train and test series

```python
train_serie=serie_inflation[:888] #serie to train our model
test_serie=serie_inflation[888:]  #serie to test our model : compare predicted values and the true values
display(train_serie)
display(test_serie)
```

We perform a Augmented Dickey-Fuller test in order to know if the original serie is stationarry

```python
#ADF test
result = adfuller(train_serie['CPIAUCSL'])
print('ADF Statistic: {}'.format(result[0]))
print('p-value: {}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))
```

From the test we clearly see that the serie is not stationnary, then we are going to study the serie of the logarithmic growth

```python
#ADF test
result = adfuller(train_serie["Log shifting"])
print('ADF Statistic: {}'.format(result[0]))
print('p-value: {}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))
```

As we can see this serie that represent the growth must be assumed stationnary with a confidence of 99%

```python
#Show stationnarity
rolling_mean=train_serie["Log shifting"].rolling(window=12).mean()
rolling_std=train_serie["Log shifting"].rolling(window=12).std()
plt.figure(figsize=(12,7))
plt.plot(train_serie["Log shifting"],label="log shifting",color='y')
plt.plot(rolling_mean,label="rolling mean")
plt.plot(rolling_std,label="rolling std")
plt.legend()
plt.grid()
plt.show()
```

The rolling mean and the rolling standard deviation are parallel to the absciss axis. It is an other way to see that this considered serie is stationary.

```python
#Plot ACF and PACF
import statsmodels.graphics.tsaplots as sgt
plt.rcParams["figure.figsize"] = 18, 5
fig, axes = plt.subplots(1, 2)
sgt.plot_acf(train_serie["Log shifting"][1:], zero = False, lags = 15, ax = axes[0])
sgt.plot_pacf(train_serie["Log shifting"][1:], zero = False, lags = 15, ax = axes[1])
plt.show()
```

from this ACF and PCF plots, we think that we must perform am ARMA model with the AR(p) term equal to q = 11 and MA(p) term equal to p = 2 on the "Log Shifting" serie 

```python
#Performing the ARIMA model from our guessed model
model=ARIMA(train_serie["Log Inflation"],order=(2,1,11))
results = model.fit()
```

```python
#Show the results
results.summary()
```

## Testing AIC and BIC coefficient for other model

```python
#Performing the ARIMA model p+1
model_p_up=ARIMA(train_serie["Log Inflation"],order=(3,1,11))
results_p_up = model_p_up.fit()
results_p_up.summary()
```

As we can see, the value for AIC and BIC are even better for this model. Then, it seems taht we must prefer ARIMA(3,1,11) to ARIMA(2,1,11). But the difference on the AIC and BIC are really small and the ARIMA(2,1,11) gives better p-value for coefficients

```python
#Performing the ARIMA model p+2
model_p_2up=ARIMA(train_serie["Log Inflation"],order=(4,1,11))
results_p_2up = model_p_2up.fit()
results_p_2up.summary()
```

As before the AIC and BIC coefficient are even better with this simulation but again the difference is not huge and the p-value computed for coefficients are again bad compare to p-values computed for the original model.
Thus, we are keeping the ARIMA(2,1,11) as our prediction model.

# Forecasting Inflation with the selected Model

```python
#Plot the Forecast of the Log values
fig, ax = plt.subplots()
ax = serie_inflation["Log Inflation"].loc['2020-01-01':].plot(ax=ax)
fig = results.plot_predict('2021-01-01', '2022-02-01', dynamic=True, ax=ax,plot_insample=False)
plt.grid()
plt.show()
```

```python
#Transforming the y-scale in order to have the original index
forecasting_val = results.forecast(14)
data = forecasting_val[0]
date_forecast = pd.date_range(start="2021-01-01",end="2022-02-02" , freq = 'MS')
forecast = pd.DataFrame(index = date_forecast , data = {'CPIAUCSL': np.exp(data)})
Forecast = train_serie["CPIAUCSL"].append(forecast['CPIAUCSL'])


# Finding the confidence interval
upper = []
lower = []
for i in range(14):
    interval = forecasting_val[2][i]
    upper.append(interval[1])
    lower.append(interval[0])
    
upper_bound = pd.DataFrame(index = date_forecast , data = {'Upper Bound': np.exp(upper)})
lower_bound = pd.DataFrame(index = date_forecast , data = {'Lower Bound': np.exp(lower)})


ARIMA_forecasting = pd.DataFrame(index = date_forecast , 
                                 data = {'Lower Bound': lower,
                                         'Upper Bound': upper,
                                         'Predictions': data})



plt.plot(Forecast[888:],color = "red" , label = "Predicted Values")
plt.plot(serie_inflation["CPIAUCSL"][888:], color = 'blue' , label = "True Data")
plt.fill_between(upper_bound.index, upper_bound['Upper Bound'], lower_bound['Lower Bound'], color='grey', alpha=0.2 , label = "Confidence Interval")
plt.legend()
plt.grid()
plt.show()
```

As we can see the red line fail to predict the boom in inlation that happens after the Covid Crisis
